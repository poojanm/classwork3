```
import pandas as pd
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines=True)
```

```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
dF = pd.DataFrame(data)
```

```
from nltk.tokenize import MWETokenizer
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk import pos_tag as pos
```

```
lmtzr = WordNetLemmatizer()
```

```
def nltkMethods(str):
    dictNltk = {}
    tk = MWETokenizer() 
    dictNltk['mwe_token'] = tk.tokenize(str.split())
    dictNltk['word_tokenize'] = word_tokenize(str)
    #WordTokenizeLemma = ' '.join([lmtzr.lemmatize(word,'v') for word in word_tokenize(str)])
    pT = nltk.pos_tag(str.split())
    dictNltk['pos_tag'] = pT

    adj = []
    for l in pT:
        if l[1] == 'JJ':
            adj.append(l[0])
    dictNltk['adjective'] = adj
    
    return dictNltk
```

```
dF['nltk']=dF['reviewText'].apply(nltkMethods)
dF.head()
```
